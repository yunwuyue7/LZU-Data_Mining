## 1.2 聚类算法简介

分类算法可以根据不同的特点和学习方式来进行区分。以下是4种常见的分类方式：

---

#### 一、基于聚类原理
最本质的分类维度，反映算法核心工作机制。

| 分类依据 | 常见算法 | 特点 |
|----------|----------|------|
| **基于划分** | KMeans、KMedoids、CLARANS | 将数据划分为预定义数量的簇，通过迭代优化簇中心（质心）和分配样本 |
| **基于层次** | AGNES（凝聚层次聚类）、DIANA（分裂层次聚类） | 构建树状聚类结构（树状图），可形成不同粒度的簇，无需预先指定簇数 |
| **基于密度** | DBSCAN、OPTICS、HDBSCAN | 根据样本密度形成任意形状的簇，能识别噪声点，无需预先指定簇数 |
| **基于模型** | 高斯混合模型（GMM）、SOM（自组织映射） | 假设数据由特定统计模型生成（如高斯分布），通过模型拟合进行聚类 |
| **基于网格** | STING、CLIQUE | 将数据空间划分为网格单元，在网格结构上进行聚类，处理速度快 |

---

#### 二、基于簇的形状
可根据数据结构选择合适算法。

| 分类依据 | 常见算法 | 特点 |
|----------|----------|------|
| **球形簇** | KMeans、高斯混合模型 | 假设簇呈凸球形分布，对非球形簇效果不佳 |
| **任意形状簇** | DBSCAN、HDBSCAN、谱聚类 | 能够发现任意形状的簇（如月牙形、环形），对簇的形状没有强假设 |
| **线性簇** | 基于密度的聚类（如DBSCAN） | 可发现线性分布的簇（如流形结构），但无专门算法，通常由密度聚类处理 |

---

#### 三、基于参数需求
根据领域知识选择调参策略。

| 分类依据 | 常见算法 | 特点 |
|----------|----------|------|
| **需指定簇数** | KMeans、层次聚类（指定切割点） | 要求用户预先指定簇的数量（K值），对K值敏感 |
| **无需指定簇数** | DBSCAN、MeanShift、Affinity Propagation | 自动确定簇的数量，但可能需要调整其他参数（如DBSCAN的eps） |
| **参数敏感** | DBSCAN（需设置eps和min_samples） | 聚类结果对参数设置敏感，需谨慎调参 |
| **参数鲁棒** | HDBSCAN、OPTICS | 对参数设置不敏感，可自动适应数据分布 |

---

#### 四、基于数据规模
针对实际应用场景的数据集选择。

| 分类依据 | 常见算法 | 特点 |
|----------|----------|------|
| **小规模数据** | 谱聚类、Affinity Propagation | 计算复杂度高（如谱聚类需计算相似矩阵），适合样本量小的场景 |
| **大规模数据** | MiniBatch KMeans、BIRCH | 通过增量学习或采样技术优化计算效率，可处理百万级样本 |
| **流式数据** | CluStream、STREAM | 支持单次遍历数据或增量更新，适用于持续到达的数据流 |
| **高维数据** | 子空间聚类（如PROCLUS） | 专门处理高维特征空间，通过特征选择降低维度灾难影响 |

---

在本次project中，共使用了10种用于分类任务的算法，这些算法的简介如下：

1. **KMeans**：基于质心的聚类算法，通过迭代优化将样本划分到K个簇中，要求预先指定簇的数量。

2. **Hierarchical Clustering**：构建树状的层次结构（树状图）来形成簇，可分为自底向上（凝聚）和自顶向下（分裂）两种方法。

3. **Gaussian Mixture Model (GMM)**：假设数据由多个高斯分布混合生成，使用期望最大化（EM）算法进行参数估计，属于软聚类。

4. **Spectral Clustering**：利用数据的相似矩阵（拉普拉斯矩阵）的特征向量进行降维，然后在低维空间进行聚类（如KMeans）。

5. **BIRCH**：适用于大规模数据集，通过构建CF树（聚类特征树）来增量地聚类，对异常值不敏感。

6. **DBSCAN**：基于密度的聚类算法，能够发现任意形状的簇，并识别噪声点，不需要预先指定簇的数量。

7. **MeanShift**：基于核密度估计的聚类算法，通过迭代寻找密度峰值点（模态）来形成簇，不需要预先指定簇的数量。

8. **OPTICS**：改进的密度聚类算法，通过计算可达距离来克服DBSCAN对参数敏感的缺点，能够发现不同密度的簇。

9. **Affinity Propagation**：通过消息传递自动确定簇的数量和代表点（exemplars），适合中小规模数据。

10. **HDBSCAN**：结合层次聚类和密度聚类的优点，能够自动确定簇的数量，对参数不敏感，适合不同密度的数据。

```python
# -*- coding: utf-8 -*-
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.metrics import adjusted_rand_score
from sklearn.cluster import (
    KMeans, DBSCAN, AgglomerativeClustering, 
    SpectralClustering, Birch, MeanShift, 
    OPTICS, AffinityPropagation, MiniBatchKMeans
)
from sklearn.mixture import GaussianMixture
from sklearn.manifold import TSNE
from sklearn.feature_selection import VarianceThreshold
from scipy.cluster.hierarchy import dendrogram, linkage
import hdbscan

# 生成超参数组合
from itertools import product  
from sklearn.metrics import adjusted_rand_score, silhouette_score  

# 解决可视化的中文字体缺失问题
from sklearn.cluster import estimate_bandwidth
import warnings
from matplotlib.font_manager import FontProperties
# 指定字体文件路径
font_path = './font/SimHei.ttf'  # 例如 'C:/Windows/Fonts/simhei.ttf'
chinese_font = FontProperties(fname=font_path)

warnings.filterwarnings("ignore", 
                        category=FutureWarning, 
                        message=".*'force_all_finite' was renamed to 'ensure_all_finite'.*")
    
# 设置绘图风格
sns.set(style="whitegrid", palette="muted")
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['figure.figsize'] = (12, 8)
```

```python
# Hyperparameter tuning configuration
TUNING_CONFIG = {
    'DBSCAN': {
        'eps': [0.3, 0.5, 0.7, 1.0],
        'min_samples': [3, 5, 10]
    },
    'MeanShift': {
        'bandwidth': [None],  # Will be estimated dynamically
        'quantile': [0.1, 0.2, 0.3],
        'bin_seeding': [True, False]
    },
    'OPTICS': {
        'min_samples': [5, 10, 15],
        'xi': [0.01, 0.05, 0.1],
        'min_cluster_size': [0.1, 0.2]
    },
    'AffinityPropagation': {
        'damping': [0.5, 0.7, 0.9],
        'preference': [None, -50, -100]
    },
    'HDBSCAN': {
        'min_cluster_size': [5, 10, 15],
        'min_samples': [1, 5, 10],
        'cluster_selection_epsilon': [0.1, 0.5]
    }
}
```

```python
class ClusteringAnalyzer:
    def __init__(self, expr_path, meta_path):
        """
        初始化聚类分析器
        :param expr_path: 基因表达数据文件路径
        :param meta_path: 元数据文件路径
        """
        self.expr_path = expr_path
        self.meta_path = meta_path
        self.X = None
        self.meta = None
        self.labels_group = None
        self.labels_das28 = None
        self.results = []
        
    def load_and_preprocess_data(self):
        """加载并预处理数据"""
        # 加载基因表达数据
        expr = pd.read_csv(
            self.expr_path, 
            sep='\t', 
            comment='!', 
            index_col=0,
            encoding='utf-8'
        ).T
        
        # 加载元数据
        meta = pd.read_csv(
            self.meta_path, 
            sep='\t', 
            quotechar='"', 
            encoding='utf-8',
            dtype=str
        )
        
        # 提取关键元数据列
        meta = meta[['geo_accession', 'samplegroup:ch1', 'das28:ch1']]
        meta.columns = ['sample_id', 'group', 'das28']
        
        # 合并数据
        expr.index.name = 'sample_id'
        merged = expr.merge(
            meta, 
            left_index=True, 
            right_on='sample_id'
        ).set_index('sample_id')
        
        # 处理缺失值
        merged['das28'] = pd.to_numeric(
            merged['das28'].replace('NA', np.nan), 
            errors='coerce'
        ).fillna(0)
        
        # 创建目标标签
        # 样本分组标签
        le_group = LabelEncoder()
        self.labels_group = le_group.fit_transform(merged['group'])
        
        # DAS28分组标签
        conditions = [
            merged['das28'] < 1,
            (merged['das28'] >= 1) & (merged['das28'] < 3),
            merged['das28'] >= 3
        ]
        choices = [0, 1, 2]  # 低, 中, 高
        self.labels_das28 = np.select(conditions, choices, default=0)
        
        # 特征矩阵
        self.X = merged.drop(['group', 'das28'], axis=1).astype(float)
        
        # 特征预处理，过滤低方差基因 (阈值设置为0.1)
        selector = VarianceThreshold(threshold=0.1)
        self.X = selector.fit_transform(self.X)
        
        # 标准化数据
        scaler = StandardScaler()
        self.X = scaler.fit_transform(self.X)
        
        # 降维 (保留95%方差)
        pca = PCA(n_components=0.95)
        self.X = pca.fit_transform(self.X)
        
        print(f"数据预处理完成, 最终维度: {self.X.shape}")
    
    def tune_hyperparameters(self, algorithm_name, param_grid):
        """Perform hyperparameter tuning for a clustering algorithm"""
        best_score = -1
        best_params = None
        best_model = None
    
        # 生成所有超参数组合
        keys, values = zip(*param_grid.items())
        param_combinations = [dict(zip(keys, v)) for v in product(*values)]
    
        for params in param_combinations:
            try:
                if algorithm_name == 'MeanShift':
                    quantile = params.pop('quantile', 0.2)
                    if 'bandwidth' in params and params['bandwidth'] is None:
                        params['bandwidth'] = estimate_bandwidth(self.X, quantile=quantile)
                
                if algorithm_name == 'HDBSCAN':
                    model_class = hdbscan.HDBSCAN  # 显式指定类
                
                # 创建模型实例
                model_class = globals()[algorithm_name]
                model = model_class(**params)
                
                if hasattr(model, 'fit_predict'):
                    clusters = model.fit_predict(self.X)
                else:
                    model.fit(self.X)
                    clusters = model.labels_
                
                n_clusters = len(np.unique(clusters[clusters != -1]))
                if n_clusters < 2 or n_clusters > 20:
                    continue
                
                # 使用轮廓系数和 ARI 进行评估
                sil_score = silhouette_score(self.X, clusters)
                ari_score = adjusted_rand_score(self.labels_group, clusters)
                combined_score = 0.7 * ari_score + 0.3 * sil_score
                
                if combined_score > best_score:
                    best_score = combined_score
                    best_params = params
                    best_model = model
                    
            except Exception as e:
                print(f"参数组合 {params} 失败: {str(e)}")
                continue
            
        # 如果没有找到有效模型，创建默认模型
        if best_model is None:
            print(f"警告: {algorithm_name} 超参数调优失败，使用默认参数")
            try:
                if algorithm_name == 'DBSCAN':
                    best_model = DBSCAN(eps=0.5, min_samples=5)
                elif algorithm_name == 'MeanShift':
                    bandwidth = estimate_bandwidth(self.X, quantile=0.2)
                    best_model = MeanShift(bandwidth=bandwidth)
                elif algorithm_name == 'OPTICS':
                    best_model = OPTICS(min_samples=10)
                elif algorithm_name == 'AffinityPropagation':
                    best_model = AffinityPropagation(damping=0.7)
                elif algorithm_name == 'HDBSCAN':
                    best_model = hdbscan.HDBSCAN(min_cluster_size=10)
                best_params = "默认参数"
            except Exception as e:
                print(f"创建默认模型失败: {str(e)}")
                # 若默认模型也失败，则返回一个简单的KMeans
                best_model = KMeans(n_clusters=3, random_state=42)
                best_params = "后备KMeans"
            
        return best_model, best_params

    def run_clustering(self):
        """Execute clustering algorithms with hyperparameter tuning"""
        algorithms = [
            ('KMeans', KMeans(n_clusters=3, random_state=42)),
            ('Hierarchical', AgglomerativeClustering(n_clusters=3)),
            ('GMM', GaussianMixture(n_components=3, random_state=42)),
            ('Spectral', SpectralClustering(
                n_clusters=3, 
                affinity='nearest_neighbors',
                n_neighbors=15,
                random_state=42
            )),
            ('BIRCH', Birch(n_clusters=3)),
            ('DBSCAN', None),  # Will be tuned
            ('MeanShift', None),  # Will be tuned
            ('OPTICS', None),  # Will be tuned
            ('AffinityPropagation', None),  # Will be tuned
            ('HDBSCAN', None)  # Will be tuned
        ]
        
        for name, model in algorithms:
            print(f"正在运行 {name} 聚类...")
        
            # 调整当前算法的超参数
            if name in TUNING_CONFIG and model is None:
                try:
                    model, best_params = self.tune_hyperparameters(name, TUNING_CONFIG[name])
                    print(f"{name} 参数: {best_params}")
                except Exception as e:
                    print(f"{name} 超参数调优失败: {str(e)}")
                    # 使用简单KMeans作为后备
                    model = KMeans(n_clusters=3, random_state=42)
                    print(f"使用后备模型: KMeans")
        
            # 确保model不为None
            if model is None:
                model = KMeans(n_clusters=3, random_state=42)
                print(f"{name} 使用后备模型: KMeans")
    
            try:
                if hasattr(model, 'fit_predict'):
                    clusters = model.fit_predict(self.X)
                else:
                    model.fit(self.X)
                    clusters = model.labels_
            except Exception as e:
                print(f"{name} 聚类失败: {str(e)}")
                # 使用随机标签作为后备
                clusters = np.random.randint(0, 3, size=len(self.X))
            
            # 评估
            ari_group = adjusted_rand_score(self.labels_group, clusters)
            ari_das28 = adjusted_rand_score(self.labels_das28, clusters)
            
            self.results.append({
                'Algorithm': name,
                'ARI (Group)': ari_group,
                'ARI (DAS28)': ari_das28,
                'Clusters': clusters,
                'Num_Clusters': len(np.unique(clusters))
            })
        
        # Convert to DataFrame
        results_df = pd.DataFrame(self.results)
        print("\n聚类分析完成!")
        return results_df    
    
    def visualize_results(self, results_df):
        """可视化聚类结果"""
        # 1. 算法性能比较
        plt.figure(figsize=(14, 6))
        plt.subplot(1, 2, 1)
        sns.barplot(x='Algorithm', y='ARI (Group)', data=results_df)
        plt.title('兰德系数 (样本分组)', fontproperties=chinese_font)
        plt.xticks(rotation=45)
        
        plt.subplot(1, 2, 2)
        sns.barplot(x='Algorithm', y='ARI (DAS28)', data=results_df)
        plt.title('兰德系数 (DAS28分组)', fontproperties=chinese_font)
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.savefig('clustering_performance.png', dpi=300)
        
        # 最佳算法可视化
        best_idx = results_df['ARI (Group)'].idxmax()
        best_algo = results_df.loc[best_idx, 'Algorithm']
        best_clusters = results_df.loc[best_idx, 'Clusters']
        
        # t-SNE降维
        tsne = TSNE(n_components=2, random_state=42)
        X_tsne = tsne.fit_transform(self.X)
        
        plt.figure(figsize=(12, 10))
        plt.subplot(2, 2, 1)
        sns.scatterplot(
            x=X_tsne[:, 0], y=X_tsne[:, 1], 
            hue=self.labels_group, palette='viridis'
        )
        plt.title('真实分组 (样本组)', fontproperties=chinese_font)
        
        plt.subplot(2, 2, 2)
        sns.scatterplot(
            x=X_tsne[:, 0], y=X_tsne[:, 1], 
            hue=self.labels_das28, palette='viridis'
        )
        plt.title('真实分组 (DAS28)', fontproperties=chinese_font)
        
        plt.subplot(2, 2, 3)
        sns.scatterplot(
            x=X_tsne[:, 0], y=X_tsne[:, 1], 
            hue=best_clusters, palette='viridis'
        )
        plt.title(f'{best_algo} 聚类结果', fontproperties=chinese_font)
        
        plt.subplot(2, 2, 4)
        linkage_matrix = linkage(self.X, method='ward')
        dendrogram(linkage_matrix, truncate_mode='level', p=3)
        plt.title('层次聚类树状图', fontproperties=chinese_font)
        plt.xlabel('样本索引', fontproperties=chinese_font)
        plt.ylabel('距离', fontproperties=chinese_font)
        
        plt.tight_layout()
        plt.savefig('clustering_visualization.png', dpi=300)
        
        # 保存结果
        results_df.to_csv('clustering_results.csv', index=False)
        print("结果已保存为 clustering_results.csv 和可视化图像")
```

```python
# 主程序
if __name__ == "__main__":
    # 初始化分析器
    analyzer = ClusteringAnalyzer(
        expr_path="./Data/GSE235508_mRNA_counts.txt",
        meta_path="./Data/GSE235508.meta.txt"
    )
    
    # 数据预处理
    analyzer.load_and_preprocess_data()
    
    # 执行聚类分析
    results = analyzer.run_clustering()
    
    # 显示结果
    print("\n聚类算法性能比较:")
    print(results[['Algorithm', 'ARI (Group)', 'ARI (DAS28)']])
    
    # 可视化结果
    analyzer.visualize_results(results)
```

警告: AffinityPropagation 超参数调优失败，使用默认参数

### 3.2.2 原因分析

| 算法                | 主要原因                                                                 |
|---------------------|--------------------------------------------------------------------------|
| **HDBSCAN**         | 代码缺陷（`globals()` 查询失败） + 高维数据下 `min_cluster_size` 过小。 |
| **DBSCAN**          | `eps` 值过小 + 高维噪声干扰。                                           |
| **OPTICS**          | 基于比例的 `min_cluster_size` 与数据集规模不匹配。                      |
| **AffinityPropagation** | 负 `preference` 值 。                                |

#### 1. 参数网格设置不当
- **DBSCAN**：  
  - `eps = [0.3, 0.5, 0.7, 1.0]` 对降维后数据尺度而言过小。  
- **OPTICS**：  
  - `xi = [0.01, 0.05, 0.1]` 范围过窄，难以覆盖复杂数据。
- **AffinityPropagation**：  
  - `preference = [None, -50, -100]`：负偏好值强制稀疏聚类（结果多为噪声）。  。
- **HDBSCAN**：   
  - `min_samples = [1, 5, 10]` 增加了噪声敏感性。

---

#### 2. 数据特性影响
- **稀疏性**：基因表达数据含大量零值/低方差特征。即使经过过滤，残余稀疏性仍干扰密度估计。  
- **簇形态**：算法假设簇为紧凑型（KMeans）或密度连通型（DBSCAN），但生物数据可能含复杂非凸结构，更适合 **谱聚类** 或 **高斯混合模型**。

## 3.3 评估指标问题

### 3.3.1 问题描述

本次聚类分析的核心目标是探索 GSE235508 转录组数据集的内在结构。该数据集包含 335 个样本，每个样本由 60,218 个基因的表达水平（CPM 值）表征。鉴于数据的高维特性，传统聚类方法面临巨大挑战，因此需要进行预处理步骤，例如特征选择（移除低方差基因）和降维（PCA）。

### 3.3.2 原因分析

#### 1. 数据高维度问题
- **问题**：  
  数据集 (`GSE235508`) 原始维度为 **60,218 个基因表达特征**，经 PCA 降维后仍有 **226 维**。基于密度的算法（DBSCAN、OPTICS、HDBSCAN）在高维空间中受“[维度灾难](https://en.wikipedia.org/wiki/Curse_of_dimensionality)”影响，距离度量失效导致密度估计不可靠。
- **影响**：  
  - `eps`（DBSCAN/OPTICS）和 `min_cluster_size`（HDBSCAN）等参数在低维数据中有效，但在高维中无法捕捉真实簇结构。  
  - 所有参数组合可能产生 **全噪声点**（无簇）或 **单一巨簇**，违反调优循环中 `2 ≤ n_clusters ≤ 20` 的条件。

---

#### 2. 评估指标问题
- **轮廓系数 (Silhouette Score)**：  
  要求至少 2 个簇且每簇含多个样本。若聚类结果全为噪声或单一簇，则无法计算。  
- **调整兰德指数 (ARI)**：  
  依赖真实标签（`labels_group`/`labels_das28`），但基因数据的自然簇可能与之不对齐。  
- 指标无定义或得负分，导致组合评分被跳过。

# 4. Summary

## 4.1 聚类模型效果对比

就各个聚类算法而言：
1. 近邻传播 (Affinity Propagation) 在样本分组上取得了最高的 ARI (0.009829)，表明其聚类结果与预定义的样本分组有微弱的一致性。
2. KMeans 和 GMM 在 DAS28 分组上表现最佳（ARI ≈ 0.0069），尽管所有算法与 DAS28 标签的匹配度都很弱。
3. DBSCAN 和 HDBSCAN 产生的聚类结果与外部标签没有一致性（ARI=0），这可能是由于过度分割或对噪声敏感所致。
4. 均值漂移 (MeanShift) 和 OPTICS 在 DAS28 分组上产生了负的 ARI 值，表明其聚类效果甚至低于随机水平。

整体来看：
- 整体 ARI 偏低：所有算法接近零的 ARI 值突显了聚类高维转录组数据的困难性。生物学异质性和技术噪音可能掩盖了清晰的组间分离。
- 算法敏感性：基于密度的方法（DBSCAN, HDBSCAN）效果不佳。
- 超参数调优：尽管进行了优化努力，像均值漂移 (MeanShift) 这样的算法仍然对带宽估计敏感，影响了簇的凝聚性。

## 4.2 后续改进

1. **维度压缩**：  
   使用 **t-SNE** 或 **UMAP** 替代 PCA，提升密度算法表现。
2. **动态参数范围**：  
   - 通过 k-距离图估计 DBSCAN/OPTICS 的 `eps`。  
   - 为 AffinityPropagation 设置基于数据相似度的 `preference`（如中位数）。
3. **算法替换**：  
   高维数据中改用 **谱聚类** 或 **高斯混合模型**。
4. **评估指标优化**：  
   当真实标签不可靠时，密度算法改用 **DBCV**（基于密度的聚类有效性指标）。
